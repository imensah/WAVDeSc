```python
# ---------- Function for clustering ----------
def cell_clustering(data, dataset_label):
    """
    data: pandas dataframe (genes x cells)
    dataset_label: string, label for plot
    """
    adata = sc.AnnData(data.T)

    # normalization + log transform
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)

    # scaling
    sc.pp.scale(adata, max_value=10)

    # PCA
    sc.tl.pca(adata, n_comps=50, svd_solver="arpack")

    # neighborhood graph + clustering
    sc.pp.neighbors(adata, n_neighbors=10, n_pcs=20)
    sc.tl.leiden(adata, resolution=0.5, key_added="clusters")

    # t-SNE
    sc.tl.tsne(adata, n_pcs=10, use_rep="X_pca")
    
    # plot
    sc.pl.tsne(adata, 
               color="clusters", 
               title=dataset_label, 
               show=True,
               colorbar_loc=None,
              save=f"_{dataset_label}_tsne.png")

    # metrics (internal, unsupervised)
    labels = adata.obs["clusters"].astype(int)
    silhouette = silhouette_score(adata.X, labels)
    db_index = davies_bouldin_score(adata.X, labels)

    return adata, {"Silhouette": silhouette, "DB Index": db_index}
```

```python
# ---------- importing data ----------
noisy = pd.read_csv("Noisy_UMI_3000.csv", index_col=0)
magic = pd.read_csv("denoised_magic.csv", index_col=0).T
saver = pd.read_csv("denoised_saver.csv", index_col=0)
enhance = pd.read_csv("denoised_enhance.tsv", sep="\t", index_col=0)
wavdesc = pd.read_csv("denoised_wavdesc_bior2.6D.csv", header=None)
dca = pd.read_csv("denoised_DCA.tsv", sep="\t", index_col=0)
```

```python
adata_noisy, metrics_noisy = cell_clustering(noisy, "Noisy (3000 cells, UMI)")
```

```python
adata_magic, metrics_magic = cell_clustering(magic, "MAGIC")
```

```python
adata_saver, metrics_saver = cell_clustering(saver, "SAVER")
```

```python
adata_enh, metrics_enh = cell_clustering(enhance, "ENHANCE")
```

```python
adata_wav, metrics_wav = cell_clustering(wavdesc, "WAVDESC")
```

```python
adata_dca, metrics_dca = cell_clustering(dca, "DCA")
```

```python
print("Noisy:", metrics_noisy)
print("MAGIC:", metrics_magic)
print("SAVER:", metrics_saver)
print("ENHANCE:", metrics_enh)
print("WAVDESC:", metrics_wav)
print("DCA:", metrics_dca)
```
