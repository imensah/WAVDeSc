```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame with the provided data for the first graph
data1 = {
    "Data": ["Noisy", "MAGIC", "SAVER", "ENHANCE", "DCA", "WAVDESC"],
    "ARI": [0.471, 0.033, 0.354, 0.332, 0.200,0.60],
    "NMI": [0.540, 0.059, 0.428, 0.369, 0.300,0.60]
}

# Create a DataFrame with the provided data for the second graph
data2 = {
    "Data": ["Original", "MAGIC", "SAVER", "ENHANCE", "DCA", "WAVDESC"],
    "ARI": [0.003, 0.000, 0.002, 0.000, 0.004, 0.004],
    "NMI": [0.005, 0.006, 0.004, 0.006, 0.005, 0.005]
}

df1 = pd.DataFrame(data1)
df1_long = pd.melt(df1, id_vars='Data', var_name='Metric', value_name='Score')

df2 = pd.DataFrame(data2)
df2_long = pd.melt(df2, id_vars='Data', var_name='Metric', value_name='Score')

# Plotting horizontally with increased label sizes
fig, axs = plt.subplots(1, 2, figsize=(18, 6.5))  # 1 Row, 2 Columns

# Plot for the first DataFrame
sns.barplot(data=df1_long, x='Data', y='Score', hue='Metric', palette=['darkgreen', 'darkred'], ax=axs[0])
axs[0].set_title('Comparison of NMI and ARI Scores for Simulated Data', fontweight='bold', fontsize=16)
axs[0].set_xlabel('Data', fontweight='bold', fontsize=14)
axs[0].set_ylabel('Scores', fontweight='bold', fontsize=14)
axs[0].grid(True, linestyle='--', color='gray', alpha=0.5)
for label in axs[0].get_xticklabels():
    label.set_fontsize(14)
    # label.set_fontweight('bold')
for label in axs[0].get_yticklabels():
    label.set_fontsize(14)
    # label.set_fontweight('bold')
legend = axs[0].legend()
plt.setp(legend.get_texts(), fontweight='bold')  

# Plot for the second DataFrame
sns.barplot(data=df2_long, x='Data', y='Score', hue='Metric', palette=['darkgreen', 'darkred'], ax=axs[1])
axs[1].set_title('Comparison of NMI and ARI Scores for PBMC Data', fontweight='bold', fontsize=16)
axs[1].set_xlabel('Data ', fontweight='bold', fontsize=14)
axs[1].set_ylabel('Scores', fontweight='bold', fontsize=14)
axs[1].grid(True, linestyle='--', color='gray', alpha=0.5)
for label in axs[1].get_xticklabels():
    label.set_fontsize(14)
    # label.set_fontweight('bold')
for label in axs[1].get_yticklabels():
    label.set_fontsize(14)
    # label.set_fontweight('bold')
legend = axs[1].legend()
plt.setp(legend.get_texts(), fontweight='bold')  

plt.tight_layout()
plt.savefig('/home/kalamix/Desktop/final-images/NMI_ARI_Comparison.png') 
plt.show()
```
